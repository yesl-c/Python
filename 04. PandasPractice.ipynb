```python
sns.scatterplot(x='longitude', y='latitude', data=data_all)
```

```python
data_01 = data_all[data_all.longitude < -122]
data_02 = data_all[(data_all.longitude < -119) & (data_tr.longitude >= -122)]
data_03 = data_all[(data_all.longitude < -118) & (data_tr.longitude >= -119)]
data_04 = data_all[data_all.longitude >= -118]
```

```python
import matplotlib.pyplot as plt

f, axes = plt.subplots(2, 2, figsize=(12,12))

axes[0,0].set_ylim([30,45])
axes[0,1].set_ylim([30,45])
axes[1,0].set_ylim([30,45])
axes[1,1].set_ylim([30,45])

sns.scatterplot(x='longitude', y='latitude', data=data_01, ax=axes[0,0])
sns.scatterplot(x='longitude', y='latitude', data=data_02, ax=axes[0,1])
sns.scatterplot(x='longitude', y='latitude', data=data_03, ax=axes[1,0])
sns.scatterplot(x='longitude', y='latitude', data=data_04, ax=axes[1,1])
```

```python
data01 = pd.read_csv("data_01.csv")
data02 = pd.read_csv("data_02.csv")
data03 = pd.read_csv("data_03.csv")
data04 = pd.read_csv("data_04.csv")
```

```python
# 새로운 컬럼을 생성
data01['group_lat'] = 1
data02['group_lat'] = 2
data03['group_lat'] = 3
data04['group_lat'] = 4
```

```python
data_a = data01.append([data02,data03,data04], ignore_index=True)
data_a.shape
```

```python
data_a.index.unique()
```

```python
# data_a 에서 data_a['group_lat'] 값의 종류별 갯수 구하기
data_a.group_lat.value_counts()
```

데이터를 그룹화하여 요약 값 확인



```python
# 그룹화
grouped = data_a.groupby('group_lat')
grouped
```

```python
grouped.mean()
```

```python
grouped.mean()[['longitude','latitude']]
```

```python
# 행정 구역 인구 데이터 평균
print('행정구역 인구 평균 :\n', grouped.mean()['population'])   
```

```python
# 소득(median_income) 데이터의 평균
print('소득 평균 : \n', grouped.mean()['median_income'])
```

```python
# 방 갯수(total_rooms) 데이터의 평균
print('방 갯수 평균 : \n', grouped.mean()['total_rooms'])
```

```python
# 세대 수(households) 데이터의 평균
print('세대 수 평균 : \n', grouped.mean()['households'])
```

그룹화한 결과를 시각화하여 확인하기

```python
fig, axes = plt.subplots(nrows=2, ncols=2)   # 2행2열의 구조
fig.set_size_inches(12,12)                 # 전체 크기

# 그룹화하기
grouped = data_a.groupby('group_lat').mean()

# 막대그래프로 확인
sns.barplot(x=grouped.index, y='population', data=grouped, ax=axes[0][0])
sns.barplot(x=grouped.index, y='median_income', data=grouped, ax=axes[0][1])
sns.barplot(x=grouped.index, y='total_rooms', data=grouped, ax=axes[1][0])
sns.barplot(x=grouped.index, y='households', data=grouped, ax=axes[1][1])

# 각 그래프에 제목 넣기
axes[0][0].title.set_text('Population')
axes[0][1].title.set_text('Median_Income')
axes[1][0].title.set_text('Total_Rooms')
axes[1][1].title.set_text('Households')
```
