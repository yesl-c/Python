```python
url2 = 'https://search.shopping.naver.com/detail/detail.nhn?nv_mid=15784793132&cat_id=50002334&frm=NVSHMDL&query=JBL+FREE+X&NaPm=ct%3Dkaoyphv4%7Cci%3Dc6bfcd8b2d9477c0c3e5dec9ad6b905ebfd0a603%7Ctr%3Dslsl%7Csn%3D95694%7Chk%3D9b8e8f7744002d4262a624b14ecc91378ae27b99'

browser = webdriver.Chrome('chromedriver')
browser.get(url2)
time.sleep(1)
```

```python
soup = BeautifulSoup(browser.page_source,'lxml')
soup.title
```

```python
dat = soup.find('ul', id='_review_list')
page_review_li = dat.find_all('li')
#page_review_li[0]
```

#### 1페이지 리뷰 가져오기

```python
all_review = []
for i in page_review_li:
    review = i.find('div', class_='atc').text
    all_review.append(review)
#print(all_review)
```

#### 2~10페이지의 리뷰 전부 가져오기
* 2페이지 : //*[@id="_review_paging"]/a[1]
* 3페이지 : //*[@id="_review_paging"]/a[2]
* 4페이지 : //*[@id="_review_paging"]/a[3]
* ...
* 10페이지 : //*[@id="_review_paging"]/a[9]

```python
for page in range(1,10):
    time.sleep(2)
    
    page_btn = '//*[@id="_review_paging"]/a[' + str(page) + ']'
    #print(page_btn)
    
    # 클릭해서 페이지 넘기기
    next_page = browser.find_element_by_xpath(page_btn)
    next_page.click()
    
    # 한 페이지의 전체 소스 가져오기
    soup = BeautifulSoup(browser.page_source,'lxml')
    
    # 해당되는 부분의 소스 가져오기
    dat = soup.find('ul', id='_review_list')
    page_review_li = dat.find_all('li')
    
    # 각각의 리뷰에 대해 정보 가져오기
    for i in page_review_li:
        review = i.find('div', class_='atc').text
        all_review.append(review)
    
    print('현재 {}페이지 완료' .format(page+1))
```
