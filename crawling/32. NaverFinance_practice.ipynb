## 국내 증시의 코스피 정보 가져오기
* url 정보 : https://finance.naver.com/sise/sise_index.nhn?code=KOSPI
* 세부적으로 원하는 데이터는 코스피지수
    * 태그명 : em
    * 추가 정보 : id='now_value'

### 절차
* html 정보를 가져온다 (urlopen)
* html 정보의 구조화 (BeatifulSoup)
* 변수/태그/함수 를 이용해서 세부적인 정보를 가져온다

```python
# 라이브러리 불러오기
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url = 'https://finance.naver.com/sise/sise_index.nhn?code=KOSPI'
page = urlopen(url)   # html 정보를 가져온다
page
```

```python
soup = BeautifulSoup(page,'lxml')
soup
```

```python
soup.title
```

```python
soup.p
```

코스피지수

```python
value = soup.find('em', id='now_value').text
```

거래량(천주)

```python
quant = soup.find('td', id='quant').text
```

거래대금(백만)

```python
amount = soup.find('td', id='amount').text
```

장중최고

```python
high_value = soup.find('td', id='high_value').text
```

장중최저

```python
low_value = soup.find('td', id='low_value').text
```

```python
print('* 코스피지수:', value)
print('* 거래량(천주):', quant)
print('* 거래대금(백만):', amount)
print('* 장중최고:', high_value)
print('* 장중최저:', low_value)
print('* 52주최고:', high_52)
print('* 52주최저:', low_52)
```

## 시황뉴스 가져오기
* 태그 정보 : span
    * 추가 정보 : class='tit'

```python
soup.find('span', class_='tit').a
```

* 태그 정보 : ul
    * 추가 정보 : class='sise_report'

```python
sise = soup.find('ul', class_='sise_report')
sise
```

```python
news = sise.find_all('a')
news
```

```python
base_url = 'http://finance.naver.com'
all_url = []

for i in news:
    #print('제목:')
    #print(i.text)
    #print('링크:')
    #print(i.attrs)   # attrs : attributes 속성
    
    one_url = base_url + i.attrs['href']
    all_url.append(one_url)
    
all_url
```

```python
url = 'http://finance.naver.com/news/news_read.nhn?article_id=0004649385&office_id=018&mode=LSS3D&type=0&section_id=101&section_id2=258&section_id3=401&page=1'
html = urlopen(url)
soup = BeautifulSoup(html,'lxml')
a = soup.find('div', id='content')
```

```python
a.text.strip()
```

```python
all_news = []
for url in all_url:
    html = urlopen(url)
    soup = BeautifulSoup(html,'lxml')
    news = soup.find('div', id='content').text.strip()
    all_news.append(news)
all_news
```

```python
import pandas as pd
news = pd.DataFrame({'news':all_news})
news.to_excel('news.xlsx', index=False)
```
