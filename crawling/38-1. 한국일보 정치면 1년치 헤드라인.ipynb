```python
%%time

driver = webdriver.Chrome('chromedriver83')

all_title=[]
for date in a:   # 날짜
    for page in range(1,3):   # 날짜별 페이지 
        driver.implicitly_wait(10)
        url = 'https://www.hankookilbo.com/News/Politics/HA01?Page={}&sDate={}'.format(page,date)
        driver.get(url)
        rand_val = randint(1,3)
        time.sleep(rand_val)
        
        soup = BeautifulSoup(driver.page_source, 'lxml')
        all_li = soup.find('ul', class_='article-list').find_all('li',class_='item')
        
        for one in all_li:
            tmp = one.find("div", class_="body").find('p')
            if tmp is not None:
                title = one.find("div", class_="body").p.text
                all_title.append(title)
            else:
                all_title.append("")
        
        #rand_val2 = randint(1,3)
        #time.sleep(rand_val2)
        
        
    print('{} 완료'.format(date))
```

```python
headline = pd.DataFrame({'headline':all_title})
headline.to_excel('headline1.xlsx', index=False)
```

```python
%%time

driver = webdriver.Chrome('chromedriver83')

all_title2=[]
for date in b:   # 날짜
    for page in range(1,3):   # 날짜별 페이지 
        driver.implicitly_wait(10)
        url = 'https://www.hankookilbo.com/News/Politics/HA01?Page={}&sDate={}'.format(page,date)
        driver.get(url)
        rand_val = randint(1,3)
        time.sleep(rand_val)
        
        soup = BeautifulSoup(driver.page_source, 'lxml')
        all_li = soup.find('ul', class_='article-list').find_all('li',class_='item')
        
        for one in all_li:
            tmp = one.find("div", class_="body").find('p')
            if tmp is not None:
                title = one.find("div", class_="body").p.text
                all_title2.append(title)
            else:
                all_title2.append("")
        
        #rand_val2 = randint(1,3)
        #time.sleep(rand_val2)
        
        
    print('{} 완료'.format(date))
```

```python
final = pd.concat([headline,headline2], axis=0)
final.to_excel('한국일보_정치면_최근1년_헤드라인.xlsx', index=False)
```
