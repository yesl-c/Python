옥션에서 'wear'로 검색했을 때 상품명 가져오기

```python
url = 'http://browse.auction.co.kr/search?keyword=wear&itemno=&nickname=&frm=hometab&dom=auction&isSuggestion=No&retry=&Fwk=wear&acode=SRP_SU_0100&arraycategory=&encKeyword=wear'
html = urlopen(url)
soup = BeautifulSoup(html, 'lxml')
soup
```

```python
soup.find('span', class_='text--title').text
```

```python
soup.find_all('span', class_='text--title')[0].text   # find_all하면 바로 .text 못함(리스트 형태로 저장되기 때문)
```

```python
all_dat = soup.find_all('span', class_='text--title')

all_title = []
for item in all_dat :
    one = item.text
    #print(one)
    all_title.append(one)
all_title[0]
```

```python
import pandas as pd

dat = pd.DataFrame({'title':all_title})
dat.head()
```

```python
dat.to_excel('web.xlsx', index=False)
```

가격 가져오기

```python
soup.find('strong', class_='text--price_seller').text
```

```python
all_p = soup.find_all('strong', class_='text--price_seller')

all_price = []
for item in all_p :
    one = item.text
    all_price.append(one)

all_dat = pd.DataFrame({'title':all_title,'price':all_price})
all_dat.to_excel('web.xlsx', index=False)
```
