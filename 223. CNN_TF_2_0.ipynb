## **TF2.0 신경망 만들기**

```
import tensorflow as tf
print(tf.__version__)
```

**데이터 셋 준비**

```
mnist = tf.keras.datasets.mnist

(x_train, y_train), (x_test, y_test) = mnist.load_data()   # 6만개와 1만개
x_train, x_test = x_train/255.0, x_test/255.0
```

```
print('학습용 데이터 : x: {}, y: {}'.format(x_train.shape, y_train.shape))
print('테스트 데이터 : x: {}, y: {}'.format(x_test.shape, y_test.shape))
```

**신경망 구성**

```
model = tf.keras.models.Sequential(
    [
     tf.keras.layers.Flatten(input_shape=(28,28)),     # Flatten : 2차원을 1차원으로
     tf.keras.layers.Dense(128, activation='relu'),    # 활성화함수 : relu (출력결과가 음수이면 0으로 만들고, 양수이면 비례한 값으로 만들어줌)
     tf.keras.layers.Dropout(0.2),                     # Dropout : 노드를 줄여주는 비율. 학습시킬때마다 20%의 노드를 랜덤하게 없애줌. 과적합 방지   
     tf.keras.layers.Dense(10, activation='softmax')   # 활성화함수 : softmax (이항분류일 경우 sigmoid, 다항분류일 경우 softmax. 예측하는 확률값으로 나타남. 확률 높은 값으로 예측)
    ]
)
```

**구성**
* sparse_categorical_crossentropy : 다중 분류 손실함수 (정수값 기준)
* categorical_crossentropy : 다중 분류 손실함수 (one-hot-encoding 기준 (예측과 실제 결과값))

```
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

**모델 훈련 및 평가**

```
model.fit(x_train, y_train, epochs=5)
```

```
model.evaluate(x_test, y_test)
```
