### 지도의 위치 찍기(Marker)

```
loc = folium.Map(location=[37.5520202,126.9161693], zoom_start=18)
folium.Marker([37.5520202,126.9161693],
              popup='yanghwajin',    # 선택 시 popup
              icon=folium.Icon(color='red', icon='star')).add_to(loc)
folium.Marker([37.5471514,126.9136486],
              popup='sungsan mid school',
              icon=folium.Icon(color='blue', icon='icon_plane')).add_to(loc)
loc
```

### CircleMarker 이용

```
m = folium.Map(location=[37.5520202,126.9161693], zoom_start=15)
folium.CircleMarker(
    location=[37.5520202,126.9161693],
    radius=100,                      # 반지름크기 지정
    color='#ffffgg',                 # 원 외곽선 색 지정
    fill_color='#fffggg',            # 원 안에 색 채우기
    popup='seoulIT',                 # 선택시 popup
    icon=folium.Icon(color='red',icon='star')    # icon을 star로 지정?
    ).add_to(m)
m
```

### Marker를 군집화 시키기 - MarkerCluster

```
import numpy as np
```

```
N = 100
data = np.array(
    [
     np.random.uniform(low=35, high=60, size=N),
     np.random.uniform(low=-12, high=30, size=N)
    ]
)
data.shape
```

```
data.T.shape
```

```
data = data.T
```

```
from folium import plugins
```

```
a=[]
for i in range(N):
  a.append(str(i))
```

```
a = [str(i) for i in range(N) if i%2==0]
a
```

```
popup_data = [str(i) for i in range(N)]
```

```
m = folium.Map([45,3], zoom_start=4, tiles='Stamen Terrain')
plugins.MarkerCluster(data,popups=popup_data).add_to(m)
# MarkerCluster: 위도, 경도 정보가 들어있는 배열 넣어주면 지도상 근처에 있는 위치끼리 묶어줌
m
```

```
m.save('map.html')
```

### GeoJson 행정 구역 데이터

```
import json
```

```
# 00. map파일 m 만들기
m = folium.Map(location=[37.5838699, 127.0565831], zoom_start=10)

# 01. 파일 불러오기
with open('seoul_muncipalities_geo.json', mode='rt', encoding='utf-8') as f:   # r: 읽기전용, t: txt, w: write
  geo = json.loads(f.read())
  print(geo)
  
# 02. folium.GeoJson() : 경계표시
folium.GeoJson(geo).add_to(m)
m
```
